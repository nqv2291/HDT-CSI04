## Bài 2. Số Lớn Thứ Hai
**Input**: Một số nguyên *n > 0* và *n* số nguyên.  
**Output**: Số lớn thứ hai trong các số đã nhập. In ra "Not found" nếu không tìm thấy.  
**Ví dụ**:  
Ví dụ 1:
- Input: n = 4, arr = [5, 4, 1, 12]
- Output: result = 5
- Giải thích: Số lớn nhất là 12, số lớn thứ hai là 5

Ví dụ 2:
- Input: n = 4, arr = [5, 4, 1, 12, 5, 12]
- Output: result = 5
- Giải thích: Có hai số cùng lớn nhất là 12, hai số cùng lớn thứ hai là 5

```python
def secondBiggest(n,arr):

    # Remove all duplicated values
    arr = set(arr)

    if len(arr)>1:  # if there are more than 1 value
        arr.remove(max(arr))
        return max(arr)
    else:
        return "Not found"

secondBiggest(2,[5,0,-1,5])
        
    
```

## Bài 3. Hình Tứ Giác
**Input**: Một số thực *a* là độ dài một cạnh của hình vuông. Hai số thực *b, c > 0* là độ dài hai cạnh kề nhau của một hình chữ nhật.  
**Output**: Chu vi và diện tích của hình vuông và hình chữ nhật.  
**Yêu cầu**: Sử dụng lại class *Rectangle* ở bài tập trên lớp. Tạo thêm class *Square* sử dụng tính kế thừa. Học viên tự chọn class được kế thừa.  
**Ví dụ**:
- Input: a = 4, b = 12, c = 2
- Output:
  + Square Perimeter: 16
  + Square Area: 16
  + Rectangle Perimeter: 28
  + Rectangle Area: 24

```python

class Rectangle:
    def __init__(self, wid, hei):
        self.wid = wid
        self.hei = hei

    def perimeter(self):
        return (self.wid + self.hei)*2
    def area(self):
        return self.wid * self.hei

class Square(Rectangle):
    def __init__(self, edge, wid=1, hei=1):
        super().__init__(wid, hei)
        self.wid = edge
        self.hei = edge   
        
hcn = Rectangle(12,2)   
hv = Square(4)

print(hv.perimeter())
print(hv.area())
print(hcn.perimeter())
print(hcn.area())
```
