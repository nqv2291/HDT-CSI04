## Bài 2. Số Lớn Thứ Hai
**Input**: Một số nguyên *n > 0* và *n* số nguyên.  
**Output**: Số lớn thứ hai trong các số đã nhập. In ra "Not found" nếu không tìm thấy.  
**Ví dụ**:  
Ví dụ 1:
- Input: n = 4, arr = [5, 4, 1, 12]
- Output: result = 5
- Giải thích: Số lớn nhất là 12, số lớn thứ hai là 5

Ví dụ 2:
- Input: n = 4, arr = [5, 4, 1, 12, 5, 12]
- Output: result = 5
- Giải thích: Có hai số cùng lớn nhất là 12, hai số cùng lớn thứ hai là 5

```python
def secondBiggest(n,arr):
    secondBiggest = 'Not found'
    biggest = arr[0]
    for i in range(0,n):
        if arr[i] > biggest:
            secondBiggest,biggest = biggest,arr[i]

    return secondBiggest   
    
```
