## Bài 1. Số Hoàn Hảo
Số hoàn hảo là một số nguyên dương mà tổng các ước nhỏ hơn nó bằng chính nó.  
**Input**: Một số nguyên *n > 0*.  
**Output**: *n* có là số hoàn hảo hay không.  
**Ví dụ**:

Ví dụ 1:  
- Input: n = 6
- Output: 6 is a perfect number
- Giải thích: Các ước của 6 nhỏ hơn chính nó là 1, 2, 3. Ta có 1+2+3 = 6

Ví dụ 2:
- Input: n = 10
- Output: 10 is NOT a perfect number
- Giải thích: Các ước của 10 nhỏ hơn chính nó là 1, 2, 5. Ta có 1+2+5 != 10

```python
# thời gian: (3n-5)--> O(n)
# không gian: n+2 --> O(n)
def so_hoan_hao(n):
    uoc = []
    for i in range(1, n):
        if n % i == 0:
            uoc.append(i)
    if sum(uoc) == n:
        return True
    return False

n = 4
if so_hoan_hao(n):
    print(f"{n} là số hoàn hảo")
else:
    print(f"{n} không là số hoàn hảo")

```

## Bài 2. Tổng Giai Thừa
**Input**: Một số nguyên *n > 0*.  
**Output**: Kết quả của biểu thức 1!+2!+3!+4!+…+N!.  
**Ví dụ**:  
- Input: n = 3
- Output: result = 9
- Giải thích: 1!+2!+3! = 1+2+6 = 9

```python
# không gian : n + 2 --> O(n)
# thời gian: 2n +2 --> O(n)
import math
def tong_giai_thua(n):
    sum = 0
    for i in range(1, n+1):
        sum += math.factorial(i)
    return sum

print(tong_giai_thua(3))
```
